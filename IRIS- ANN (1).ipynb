```python
import numpy as np
import pandas as pd
from tensorflow import keras
from keras.utils import np_utils
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import seaborn as sb
from keras.models import Sequential
from keras.layers import Dense
import sklearn
from sklearn import metrics
from sklearn import preprocessing
print("hello world")
```

```python
iris= pd.read_csv(r'C:\Users\aadit\OneDrive\Desktop\IRIS dataset\Iris.csv')
x=iris[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]].values
x=preprocessing.scale(x)
y=iris["Species"].values
enc=LabelEncoder()
Y=pd.get_dummies(enc.fit_transform(y)).values

print(Y.shape)
print(y.shape)
x_train, x_test, y_train,y_test=train_test_split(x,Y,test_size=0.2,stratify=Y)

```

```python
model=Sequential()
model.add(Dense(10, input_dim=4, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(3, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=100, batch_size=10)
```

```python
print(model.evaluate(x_test, y_test))
```

```python
expected=y_test
expected=np.argmax(expected,axis=1)
prediction=model.predict(x_test)
prediction=np.argmax(prediction,axis=1)

print(metrics.classification_report(expected,prediction))
print(metrics.confusion_matrix(expected,prediction))
```
